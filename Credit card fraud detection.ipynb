```python
#Under_Sampling
#Sample dataset containing similar distribution of normal and fraudulent transactions
#number of fraudulent transactions
legit_sample=legit.sample(n=492)
```

```python
#concartinating 2 Dataframs
```

```python
new_dataset=pd.concat([legit_sample,fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Split the data into X and Y
X=new_dataset.drop(['Class'], axis=1)
Y=new_dataset['Class']
```

```python
X.head()
```

```python
print(Y)
```

```python
#split data into training and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#creating a model
model=LogisticRegression()
```

```python
#Fit/train model on training data
model.fit(X_train,Y_train)
```

```python
#preview predicted data
model.predict(X_test)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy score is:', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy score is:', testing_data_accuracy)
```

```python
#Saving model
dump(model,'Credit_card_fraud_detection.joblib')
```
